```python
given_text = input("Please provide the text: ")
Analysis = input(
    """Given your text, which analysis would you like to perform?
Please select from the following options:
A) Total Word Count        I) Longest Word in The Text     Q) Count Syllables
B) Total Unique Word Count J) Sentence Count               R) Flesch Reading Ease Score
C) Most Frequent Words     K) Vowel Consonant Ratio        S) Flesch-Kincaid Grade Level
D) Words > X Letters       L) Lexical Diversity            T) Keyword in Context (kwic)
E) Average Word Length     M) Find Anagrams        
F) Least Frequent Word     N) Words with Specific Substr
G) Words Starting Letter   O) % of Stop Words
H) Words Ending Letter     P) Find Palindromes

Your choice: """
)

Analysis = Analysis.upper()

def total_word_count(text):
    # return the total number of words in the text
    return len(text.split())

def total_unique_word_count(text):
    # return the count of unique word in the text
    words = text.split()
    words = len(set(words))
    return words

def most_frequent_words(text,n=10):
    # return the most frequent words and their frequencies
    words = text.split()
    word_frequency = {word : words.count(word) for word in set(words)}
    sorted_word = sorted(word_frequency.items(), key = lambda x: x[1], reverse = True)
    return sorted_word[:n]

def words_with_more_then_x_letter(text,n):
    # returns a list of words that have more than 'n' letters
    words = text.split()
    words = [word for word in words if len(word)>int(n)]
    return words

def average_word_length(text):
    # calculate the average length of words in the text
    words = text.split()
    total_length = sum(len(word) for word in words)
    return total_length/len(words)

def least_frequent_word(text):
    # finds and returns the least frequent word in the text
    words = text.split()
    words_dict = {word: words.count(word) for word in words}
    sorted_dict = sorted(words_dict.items(),key = lambda x : x[1], reverse = False)
    return sorted_dict[0]

def words_starting_with_specific_letter(text,letter):
    # Finds words that start with a specific letter
    words = text.split()
    collected_list = [word for word in words if word[0]==letter]
    removing_duplicate = []
    [removing_duplicate.append(word) for word in collected_list if word not in removing_duplicate]
    return removing_duplicate

def words_ending_with_specific_letter(text,letter):
    # Finds words that end with a specific letter
    words = text.split()
    collected_list = [word for word in words if word[-1]==letter]
    removing_duplicate = []
    [removing_duplicate.append(word) for word in collected_list if word not in removing_duplicate]
    return removing_duplicate
    
def longest_word_in_the_text(text):
    # identifies the longest word in the text
    words = text.split()
    longest_word=words[0]
    [longest_word:= word for word in words if len(word)>len(longest_word)]
    return longest_word

def sentence_count(text):
    # count the number of sentence in the text
    count = 0 
    for char in text:
        if char in ['.','?',"!"]:
            count+=1
    return count

def vowel_consonant_ratio(text):
    # calculate the ratio of vowels to consonants in the text
    vowel = "AEIOUaeiou"
    consonant = "BCDFGHJKLMNPQRSTVWXYZbcdfghjklmnpqrstvwxyz"
    num_vowel = sum(1 for char in text if char in vowel)
    num_consonant = sum(1 for char in text if char in consonant)
    print("No of Vowel: ",num_vowel)
    print("No of consonant: ",num_consonant)
    if num_consonant ==0:
        return "No consonants to calculate a ratio"
    return num_vowel/num_consonant

def lexical_diversity(text):
    # computes the lexical diversity of the text
    words = text.split(text)
    unique_word = len(set(words))
    total_words = len(words)
    return unique_word/total_words

def find_anagrams(text,word):
    # find anagrams of a given word within the text
    text = "".join(char for char in text if char.isalnum() or char.isspace()).lower()
    words = text.split()
    sorted_word = sorted(word)
    anagrams = [candidate for candidate in words if sorted(candidate)== sorted_word and
              candidate != word]
    anagrams =list(set(anagrams))
    return anagrams

def words_containing_substring(text,substring):
    # identifies words that contain a specific substring
    text = text.lower()
    substring = substring.lower()
    words = text.split()
    return [word for word in words if substring in word]

def percentage_of_stop_words(text):
    # calculated the percentage of stop word in the text
    words = text.lower().split()
    stop_words = {'a', 'an', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 
    'for', 'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 
    'on', 'or', 'such', 'that', 'the', 'their', 'then', 'there', 
    'these', 'they', 'this', 'to', 'was', 'will', 'with'}
    count = sum(1 for word in words if word in stop_words)
    return (count/len(words))*100

def find_palindromes(text):
    #find the palindromic word in the text
    text = ''.join(char.lower() for char in text if char.isalnum() or char.isspace())
    words = text.split()
    palindromes = {word for word in words if word == word[::-1] and len(word)>1}
    return list(palindromes)

def count_syllables(text):
     # count the syllable of each word in the text
    words = text.lower().split()
    syllable = {}
    vowels = "aeiou"
    for word in words:
        syllables=0
        if word[0] in vowels:
            syllables +=1
        for index in range(1,len(word)):
            if word[index] in vowels and word[index-1] not in vowels:
                syllables +=1
        if word.endswith("e") and len(word)>1 and word[-2] not in vowels:
            syllables-=1
        if word.endswith("le") and len(word)>2 and word[-3] not in vowels:
            syllables+=1
        if syllables ==0:
            syllables=1
        syllable[word] = syllables
    return syllable

def flesh_reading_ease(text):
    # calculate the flesh reading ease score for the text
    while True:
        sentences = text.count('.') + text.count('!') + text.count('?')
        if sentences == 0:
            print("Please enter a valid text ending with '.','!','?'")
            text = input("Please provide the text: ")
            continue
        words = len(text.split())
        if sentences == 0 or words ==0:
            print("Enter a valid text please.")
            continue
        word_to_syllable = count_syllables(text)
        syllables = sum(word_to_syllable.values())
        try:
            score = 206.835 - 1.015 * (words/sentences) - 84.6 * (syllables/words)
            return round(score,2)
        except ZeroDivisionError:
            print("The text must contain at least one sentence.")
            continue

def flesh_kincaid_grade_level(text):
    # calculate the Flesh-Kincaid grade level for the text
    while True:
        sentences = text.count(".") + text.count("!") + text.count("?")
        if sentences ==0:
            print("Please enter a valid text ending with '.','!','?'")
            text = input("Please provide the text: ")
            continue
        words = len(text.split())
        if sentences ==0 or words ==0:
            print("Enter a valid text please.")
            continue
        word_to_syllable = count_syllables(text)
        syllables = sum(word_to_syllable.values())
        try:
            score = 0.39 * (words/sentences) + 11.8 * (syllables/words) - 15.59
            return round(score,2)
        except ZeroDivisionError:
            print("The Text must contain at least one sentence")
            continue

def kwic(text,keyword,context_size=5):
    # provide the keyword in context analysis for the text
    text = text.lower()
    keyword = keyword.lower()
    words = text.split()
    keyword_indices = [i for i,word in enumerate(words) if word == keyword]
    contexts = []
    for index in keyword_indices:
        start = max(0,index-context_size)
        end = index + context_size + 1
        context = words[start:index],words[index],words[index+1:end]
        contexts.append((' '.join(context[0]), context[1],' '.join(context[2])))
    return contexts

valid_options = set("ABCDEFGHIJKLMNOPQRSTU")

while True:
    Analysis = Analysis.upper()
    if Analysis in valid_options:
        pass
    if Analysis.lower() == "break":
        print("Exiting the Program")
        break
    else:
        print("Invalid input. Please select from the options A to T.")
    if Analysis == "A":
        print("Total Word Count:", total_word_count(given_text))
    elif Analysis == "B":
        print("Total Unique word Count:",total_unique_word_count(given_text))
    elif Analysis == "C":
        print("Most Frequent Words:", most_frequent_words(given_text))
    elif Analysis == "D":
        n = input("Enter the minimum number of letters for the words you'd like to see:\n")
        print('\n')
        print(f"Words with more the {n} letters:",words_with_more_then_x_letter(given_text,n))
    elif Analysis == "E":
        print("Average word length:",average_word_length(given_text))
    elif Analysis == "F":
        print("Least Frequent Word:", least_frequent_word(given_text))
    elif Analysis == "G":
        letter = input("Please input the required letter: ")
        print("Words Starting with a specific Letter", words_starting_with_specific_letter(given_text,letter))
    elif Analysis == "H":
        letter = input("Please input the required letter: ")
        print("Words Ending with a specific Letter", words_ending_with_specific_letter(given_text,letter)) 
    elif Analysis == "I":
        print("Longest Word in the Text",longest_word_in_the_text(given_text))
    elif Analysis == "J":
        print("Total No of sentence in the text:",sentence_count(given_text))
    elif Analysis == "K":
        print("Vowel Consonant Ratio is:",vowel_consonant_ratio(given_text))
    elif Analysis =="L":
        print("Lexical Diversity Percentage:",lexical_diversity(given_text))
    elif Analysis == "M":
        user_word = input("Enter a word to find its anagram in the text: ")
        print("Anagrams Found:", find_anagrams(given_text,user_word))
    elif Analysis == "N":
        substring = input("Enter the substring to search for: ")
        print("Words containing the substring: ",words_containing_substring(given_text,substring))
    elif Analysis == "O":
        print("Percentage of stop words: ",percentage_of_stop_words(given_text))
    elif Analysis =="P":
        print("List of Palindromic Words: ", find_palindromes(given_text))
    elif Analysis =="Q":
        print("Syllable count for each word:",count_syllables(given_text))
    elif Analysis == "R":
        print("Flesh Reading Ease Score: ",flesh_reading_ease(given_text))
    elif Analysis == "S":
        print("Flesch-Kincaid Grade Level",flesh_kincaid_grade_level(given_text))
    elif Analysis == "T":
        keyword = input("Enter the keyword please: ")
        context_size = int(input("Enter the context size: "))
        print("Keyword in Context(Kwic)",kwic(given_text,keyword,context_size))
        
    Analysis = input(
    """Do you wish to continue with another analysis from the list or exit?
Please write 'break' to exit.
A) Total Word Count           I) Longest Word in The Text    Q) Count Syllables
B) Total Unique Word Count    J) Sentence Count              R) Flesch Reading Ease Score
C) Most Frequent Words        K) Vowel Consonant Ratio       S) Flesch-Kincaid Grade Level
D) Words with More Than X     L) Lexical Diversity           T) Keyword in Context (kwic)
E) Average Word Length        M) Find Anagrams               U) Input New Text
F) Least Frequent Word        N) Words with Specific Substr
G) Words Starting Letter      O) Percentage of Stop Words
H) Words Ending Letter        P) Find Palindromes

Your choice: """
)
    Analysis = Analysis.strip().upper()

    if Analysis == "U":
        given_text = input("Please enter your new text:\n")
        continue
    if Analysis.lower() == "break":
        print(" Exiting the program")
        break
```
